### Monthly HDD - graph

Comparison of heating demands in the most populated cities in Quebec province. 

http://climate.weather.gc.ca/index_e.html

```python
%matplotlib inline
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SIZE)                # controls default text sizes
plt.rc('axes', titlesize=SIZE)           # fontsize of the axes title
plt.rc('axes', labelsize=SIZE)           # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)          # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)          # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
plt.style.use('seaborn-deep')
```

```python
hdd_qc=pd.read_csv('monthly HDD/monthly HDD by city.csv',sep=';',index_col='date',parse_dates=True,dayfirst=True)
```

```python
hdd_qc.plot(figsize=(5.6,2.5))
a=plt.gca()
L=plt.legend(loc='upper center',ncol=4,framealpha=0.5)
a.set_ylabel('monthly heating degree days\n (celsius degrees)')
plt.tight_layout()
#plt.savefig('HDD by city.png',dpi=800) (for the article)
```

The trends are actually quite similar. Even if some cities are colder than other the availability factor (i.e. output/capacity) of heating equipment is likely to be similar for most of the heat demand in Quebec.
